```python
#making dates into datetime format
main_df["DATE_FDA_FIRST_RECEIVED_REPORT"] = main_df['DATE_FDA_FIRST_RECEIVED_REPORT'].astype('datetime64')
main_df["DATE_EVENT"] = main_df['DATE_EVENT'].astype('datetime64')
```

```python
#splitting product code into the respective numbers and letters 
main_df[['CLASS']] = main_df['PRODUCT_CODE'].str.split('\\d+', n=1) #letters
main_df[['P_CODE']] = main_df['PRODUCT_CODE'].str.split('\\D+', n=1) #numbers

main_df['P_CODE'] = main_df['P_CODE'].str[0]
main_df['P_CODE'] = main_df['P_CODE'].astype('float32')

main_df['CLASS'] = main_df['CLASS'].str[1]
```

```python
#splitting report ID such that we only get the number, as most of them are formatted like that anyway
main_df['split_code'] = main_df['REPORT_ID'].str.split('-')
main_df['REPORT_ID'] = main_df['split_code'].str[2]

#saving the date part of the report ID for columns that don't have a date-- 
#setting default to january 1st of that year

def estimate_year(row):
    if pd.isnull(row['DATE_EVENT']):
    #if .isnull():
        #if not row['split_code'] == "":
        year_arr = row['split_code']
        year = year_arr[0]
        if int(year) >= 1990 and int(year) < 2024:
            return pd.to_datetime(year+'-01-01', format='%Y-%m-%d')
    else:
        return row['DATE_EVENT']

main_df['DATE_EVENT'] = main_df.apply(estimate_year, axis=1)
main_df["REPORT_ID"] = main_df['REPORT_ID'].astype('float32')
```

```python
#converting age into all years, so they aren't separate units 

def convert_age(row):
    if row['AGE_UNITS'] == 'month(s)':
        return (row['PATIENT_AGE']/12)
    if row['AGE_UNITS'] == 'day(s)':
        return (row['PATIENT_AGE']/365)
    if row['AGE_UNITS'] == 'week(s)':
        return (row['PATIENT_AGE']/ 52)
    else:
        return row['PATIENT_AGE']

main_df['PATIENT_AGE'] = main_df.apply(convert_age, axis=1)
main_df = main_df.drop(columns=['AGE_UNITS'], axis=1)
```

```python
#dropping a column of NaN values, split values, Product_code
main_df = main_df.drop(columns =['Unnamed: 12', 'split_code', 'PRODUCT_CODE'], axis=1)
main_df = main_df.drop(main_df[main_df['PRODUCT'] == 'EXEMPTION 4'].index)
```

```python
#severity across every category
severity_map = {
    'Allergic Reaction': 2,
    'Congenital Anomaly': 8,
    'Disability': 8,
    'Hospitalization': 6,
    'Life Threatening': 8,
    'Other Serious Outcome': 4,
    'Other Serious or Important Medical Event': 4,
    'Required Intervention': 4,
    'Visited Emergency Room': 4,
    'Visited a Health Care Provider': 2,
    'Death': 10,
    'Other Outcome': 1,
    'Injury': 3
}
```

```python
#assign each to a severity score

def calculate_severity_score(row):
    outcomes = row['CASE_OUTCOME']
    curr_score = 0
    for symptom in severity_map:
        if symptom in outcomes and severity_map[symptom] > curr_score:
            curr_score = severity_map[symptom]
    return curr_score
            
            

main_df['SEVERITY SCORE'] = main_df.apply(calculate_severity_score, axis=1)
```

```python
average_severity = main_df.groupby('DESCRIPTION').mean()['SEVERITY SCORE']
```

```python
plt.figure(figsize = (30,15))
sns.barplot(x=average_severity.index, y = average_severity.values)
plt.title("Severity Scores Across Each Category", fontsize=60)
plt.xticks(rotation=89, fontsize = 20)
plt.yticks(fontsize=20)
plt.show()
```
