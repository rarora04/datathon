```python
#converted ages 
def convert_age(row):
    if row['AGE_UNITS'] == 'month(s)':
        return (row['PATIENT_AGE']/12)
    if row['AGE_UNITS'] == 'day(s)':
        return (row['PATIENT_AGE']/365)
    if row['AGE_UNITS'] == 'week(s)':
        return (row['PATIENT_AGE']/ 52)
    else:
        return row['PATIENT_AGE']

main_df['PATIENT_AGE'] = main_df.apply(convert_age, axis=1)
#main_df['AGE_UNITS'] = main_df.replace({'month(s)': 'years'})
```

```python
severity_map = {
    'Allergic Reaction': 2,
    'Congenital Anomaly': 8,
    'Disability': 8,
    'Hospitalization': 6,
    'Life Threatening': 8,
    'Other Serious Outcome': 4,
    'Other Serious or Important Medical Event': 4,
    'Required Intervention': 4,
    'Visited Emergency Room': 4,
    'Visited a Health Care Provider': 2,
    'Death': 10,
    'Other Outcome': 1,
    'Injury': 3
}
```

```python
def calculate_severity_score(row):
    outcomes = row['CASE_OUTCOME']
    curr_score = 0
    for symptom in severity_map:
        if symptom in outcomes and severity_map[symptom] > curr_score:
            curr_score = severity_map[symptom]
    return curr_score
           
           

main_df['SEVERITY SCORE'] = main_df.apply(calculate_severity_score, axis=1)
```

```python
# Filter for older men and sort by severity score
young_men = main_df[(main_df['SEX']=='Male') & (main_df['PATIENT_AGE'] <= 48)].sort_values(by='SEVERITY SCORE', ascending=False)
young_women = main_df[(main_df['SEX']=='Female') & (main_df['PATIENT_AGE'] <= 53)].sort_values(by='SEVERITY SCORE', ascending=False)
old_men = main_df[(main_df['SEX']=='Male') & (main_df['PATIENT_AGE'] > 48)].sort_values(by='SEVERITY SCORE', ascending=False)
old_women = main_df[(main_df['SEX']=='Female') & (main_df['PATIENT_AGE'] > 53)].sort_values(by='SEVERITY SCORE', ascending=False)
```

```python
#Top 3 issues for young men
above_5_ym = young_men[young_men['SEVERITY SCORE'] > 5]
top_3_issues_ym = above_5['DESCRIPTION'].value_counts().nlargest(3)
top_3_issues_ym
```
